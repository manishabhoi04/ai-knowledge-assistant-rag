```python
all_text = ""

for page in doc:
    all_text += page.get_text() + "\n"

print("Total Characters:", len(all_text))
```

```python
chunk_size = 1000

chunks = []

for i in range(0, len(all_text), chunk_size):
    chunk = all_text[i:i + chunk_size]
    chunks.append(chunk)

print("Total Chunks:", len(chunks))
```

```python
print(chunks[20])
```

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer("BAAI/bge-small-en-v1.5")
```

```python
embedding = model.encode(chunks[20])

print(type(embedding))
print(len(embedding))
```

```python
print(embedding[:20])
```

```python
chunk_embeddings = model.encode(chunks)

print(type(chunk_embeddings))
print(chunk_embeddings.shape)
```

```python
query = "What is probability?"

query_embedding = model.encode(query)

print(query_embedding.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

scores = cosine_similarity(
    [query_embedding],
    chunk_embeddings
)

print(scores.shape)
```

```python
import numpy as np

best_index = np.argmax(scores)

print("Best Chunk Index:", best_index)
```

```python
print(chunks[best_index])
```

```python
top_k = 3

top_indices = np.argsort(scores[0])[-top_k:][::-1]

print(top_indices)
```

```python
for idx in top_indices:
    print("="*80)
    print("Chunk:", idx)
    print(chunks[idx][:500])
```

```python
def ask_rag(question):

    # Convert question to embedding
    query_embedding = model.encode(question)

    # Retrieve top 3 chunks
    results = collection_full.query(
        query_embeddings=[query_embedding.tolist()],
        n_results=3
    )

    retrieved_chunks = results["documents"][0]

    # Build context
    context = "\n\n".join(retrieved_chunks)

    # Create prompt
    prompt = f"""
    Use the following context to answer the question.

    Context:
    {context}

    Question:
    {question}

    Answer:
    """

    # Send to Ollama
    response = requests.post(
        "http://localhost:11434/api/generate",
        json={
            "model": "llama3.2",
            "prompt": prompt,
            "stream": False
        }
    )

    return response.json()["response"]
```

```python
answer = ask_rag("What is probability theory?")

print(answer)
```
